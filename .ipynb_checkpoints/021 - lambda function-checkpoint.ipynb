```python
# S = ut + 1/2 a*t**2
def distance_eqn(u,a):
    return lambda t: u*t + ((1/2)*a*t**2)

dist = distance_eqn(5,10)
print(dist(20))
```

```python
print(distance_eqn(5,10)(20))
```

```python
time = []
dist = []

for t in range(0, 1000, 2):
    d = distance_eqn(5,10)(t)
    time.append(t)
    dist.append(d)

from matplotlib import pyplot as plt
plt.plot(time, dist)
```
