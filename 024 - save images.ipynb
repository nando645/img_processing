```python
from skimage import filters, io, img_as_ubyte
from matplotlib import pyplot as plt
```

```python
img = io.imread('images/Osteosarcoma_01.tif')
```

```python
gaussian_img = filters.gaussian(img, sigma=10)
```

```python
plt.imshow(img)
```
